```python
import pandas as pd
import glob
import matplotlib.pyplot as plt
%matplotlib inline
# list of file names
files = glob.glob('/Users/santi/Desktop/Winter 2019/B ME 450 (Ocean Engineering)/Final Project, Surface inshore/*.csv')

# concat csvs
dfs = []
for file in files:
  dfs.append(pd.read_csv(file))
df = pd.concat(dfs)

# plot

files2 = glob.glob('/Users/santi/Desktop/Winter 2019/B ME 450 (Ocean Engineering)/Near-surface/*.csv')

# concat csvs
dfs2 = []
for file in files2:
  dfs2.append(pd.read_csv(file))
df2 = pd.concat(dfs2)

files3 = glob.glob('/Users/santi/Desktop/Winter 2019/B ME 450 (Ocean Engineering)/Seafloor/*.csv')

# concat csvs
dfs3 = []
for file in files3:
  dfs3.append(pd.read_csv(file))
df3 = pd.concat(dfs3)

# plot

plt.plot(df.time, df.temp,label='1m')
plt.plot(df2.time, df2.temp,label='7m')
plt.plot(df3.time, df3.temp,label='29m')
plt.title('Temperature at different depths')
plt.xlabel('Time as recorded by CTD')
plt.ylabel('Temperature in Celsius')
plt.legend()
plt.show()
```
